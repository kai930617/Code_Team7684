```python
target_player = '孫穎莎(RSH)'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm



plt.rcParams['font.family'] = 'Microsoft YaHei'  
plt.rcParams['axes.unicode_minus'] = False  

# 讀取CSV檔案
file_path = f"converted_data/{target_player}_data_converted.csv"
df = pd.read_csv(file_path)

# 要繪製長條圖的欄位
columns_to_plot = ['正反手', '己方落點', '對方落點', '技術', '速度', '旋轉', '長短', '變線', "階段"]

# 設置繪圖大小
plt.figure(figsize=(20, 15))

# 繪製長條圖
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(3, 3, i)
    
    # 計算每個類別的次數
    value_counts = df[column].value_counts()
    
    # 繪製長條圖
    bars = plt.bar(value_counts.index, value_counts.values, color='skyblue', edgecolor='black')
    
  
    plt.title(f'{column} 分佈', fontsize=24)  
    plt.xlabel('類別', fontsize=22)  
    plt.ylabel('次數', fontsize=22)  
    
  
    plt.xticks(rotation=45, ha='right', fontsize=20)
    
   
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.0f}', 
                 ha='center', va='bottom')


plt.tight_layout()

plt.suptitle(f'{target_player} 特徵技術分布圖', fontsize=28, y=1.02)

plt.show()
```
